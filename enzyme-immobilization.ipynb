```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
free_data = {
    "enzyme_type": ["Free"] * 5,
    "substrate_mg_ml": [35.714, 71.428, 107.142, 142.857, 178.571],
    "od_540": [0.3757, 0.6488, 0.8879, 1.1472, 1.1320],
    "glucose_mg_ml": [0.5436, 0.8694, 1.1547, 1.4640, 1.4459],
    "glucose_mmol_ml": [3.02, 4.83, 6.41, 8.13, 8.03],
    "velocity_V": [0.302, 0.483, 0.641, 0.813, 0.803]
}

df_free = pd.DataFrame(free_data)
df_free
```

```python
immobilized_data = {
    "enzyme_type": ["Immobilized"] * 5,
    "substrate_mg_ml": [35.714, 71.428, 107.428, 142.857, 178.571],
    "od_540": [0.0007, 0.0075, 0.0062, 0.0090, 0.0103],
    "glucose_mg_ml": [0.0103, 0.1043, 0.1028, 0.1061, 0.1077],
    "glucose_mmol_ml": [0.0572, 0.5792, 0.5711, 0.5894, 0.5983],
    "velocity_V": [0.0057, 0.0579, 0.0571, 0.0589, 0.0598]
}

df_imm = pd.DataFrame(immobilized_data)
df_imm
```

```python
df = pd.concat([df_free, df_imm], ignore_index=True)
df
```

```python
df["inv_velocity_1_V"] = 1 / df["velocity_V"]
df["inv_substrate_1_S"] = 1 / df["substrate_mg_ml"]

df
```

```python
mm_df = df[["enzyme_type", "substrate_mg_ml", "velocity_V"]]
mm_df
```

```python
lb_df = df[["enzyme_type", "inv_substrate_1_S", "inv_velocity_1_V"]]
lb_df
```

```python
plt.figure()

for enzyme, group in df.groupby("enzyme_type"):
    plt.scatter(group["substrate_mg_ml"], group["velocity_V"], label=enzyme)

plt.xlabel("Substrate concentration (mg/ml)")
plt.ylabel("Reaction velocity (V)")
plt.title("Free vs Immobilized Enzyme Kinetics")
plt.legend()
plt.show()
```

```python
plt.figure()

for enzyme, group in df.groupby("enzyme_type"):
    plt.scatter(group["inv_substrate_1_S"], group["inv_velocity_1_V"], label=enzyme)

plt.xlabel("1 / Substrate (1/mgml)")
plt.ylabel("1 / Velocity (1/V)")
plt.title("LineweaverBurk Plot")
plt.legend()
plt.show()
```

### Interpretation: LineweaverBurk Plot

The LineweaverBurk plot (1/V versus 1/S) provides a linearized representation of enzyme 
kinetics, allowing comparison of kinetic parameters between free and immobilized enzymes.

The immobilized enzyme shows a higher y-intercept (1/Vmax) compared to the free enzyme, 
indicating a lower maximum reaction velocity. Additionally, the steeper slope observed 
for the immobilized enzyme suggests an increased apparent Km, reflecting reduced 
substrate affinity.

These changes arise due to mass transfer resistance, restricted diffusion of substrate 
to the active site, and possible conformational changes of the enzyme upon immobilization.

Thus, the LineweaverBurk analysis confirms the reduction in catalytic efficiency caused 
by enzyme immobilization.
